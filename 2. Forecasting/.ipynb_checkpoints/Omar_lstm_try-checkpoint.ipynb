```python
import holidays

def add_time_features(df):
    cet_index = df.index
    df["month"] = cet_index.month
    df["weekday"] = cet_index.weekday
    df["hour"] = cet_index.hour
    #df["year"] = cet_index.year
    return df

def add_holiday_features(df):
    de_holidays = holidays.France()
    cet_dates = pd.Series(df.index, index=df.index)
    df["holiday"] = cet_dates.apply(lambda d: d in de_holidays)
    df["holiday"] = df["holiday"].astype(int)
    return df

def add_lag_features(df, col="conso_global"):
    for n_hours in range(24, 49):
        shifted_col = df[col].shift(n_hours, "h")
        shifted_col = shifted_col.loc[df.index.min(): df.index.max()]
        label = f"{col}_lag_{n_hours}"
        df[label] = np.nan
        df.loc[shifted_col.index, label] = shifted_col
    return df

def add_all_features(df, target_col="conso_global"):
    df = df.copy()
    df = add_time_features(df)
    #df = add_holiday_features(df)
    #df = add_lag_features(df, col=target_col)
    return df
```

# window 1h: 1h jump

```python
next_hour = 12
window = 12

x_train, y_train = [], []


for i in range(window, train_data.shape[0]-window, window):
    x_train.append(train_data_scaled[i-window:i])
    y_train.append(train_data_scaled[i:i+window])

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape, y_train.shape
```

```python
next_hour = 12
window = 12

x_test, y_test = [], []


for i in range(window, test_data.shape[0]-window, window):
    x_test.append(test_data_scaled[i-window:i])
    y_test.append(test_data_scaled[i:i+window])

x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
x_test.shape, y_test.shape
```

```python
# Define some common parameters

input_shape= x_train.shape[1],x_train.shape[2]
loss = tf.keras.losses.MeanSquaredError()
metric = [tf.keras.metrics.RootMeanSquaredError()]
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
              lambda epoch: 1e-4 * 10**(epoch / 10))
early_stopping = tf.keras.callbacks.EarlyStopping(patience=1, verbose=1)
```

```python
def plot_model_rmse_and_loss(history):
    
    # Evaluate train and validation accuracies and losses
    
    train_rmse = history.history['root_mean_squared_error']
    #val_rmse = history.history['val_root_mean_squared_error']
    
    train_loss = history.history['loss']
    #val_loss = history.history['val_loss']
    
    # Visualize epochs vs. train and validation accuracies and losses
    
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 2, 1)
    plt.plot(train_rmse, label='Training RMSE')
    #plt.plot(val_rmse, label='Validation RMSE')
    plt.legend()
    plt.title('Epochs vs. Training and Validation RMSE')
    
    plt.subplot(1, 2, 2)
    plt.plot(train_loss, label='Training Loss')
    #plt.plot(val_loss, label='Validation Loss')
    plt.legend()
    plt.title('Epochs vs. Training and Validation Loss')
    
    plt.show()
```

```python
tf.keras.backend.clear_session()

model = Sequential()

model.add(LSTM(24, input_shape=input_shape, 
         return_sequences=True))
# model.add(Bidirectional(LSTM(12, activation='relu', return_sequences=True)))
# model.add(Bidirectional(LSTM(24, activation='relu')))
model.add(Flatten())


# model.add(Dense(units = 200, activation='relu'))
          
model.add(Dropout(0.1))

model.add(Dense(units = window))


model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
                   'multivariate_lstm.h5', monitor=('val_loss'), save_best_only=True)

optimizer = tf.keras.optimizers.Adam(lr=0.001) #lr=6e-3, amsgrad=True

model.compile(loss=loss, optimizer=optimizer, metrics=metric)
```

```python
history = model.fit(x_train, y_train, epochs=20)
```

```python
plot_model_rmse_and_loss(history)
```

```python
y_test = y_test[:,:].reshape(y_test.shape[0], y_test.shape[1])
```

```python
y_test.shape
```

```python
forecast = model.predict(x_test)
# lstm_forecast = scaler_y.inverse_transform(forecast)

rmse_lstm = np.sqrt(mean_squared_error(y_test,forecast))
print('RMSE of hour-ahead electricity price LSTM forecast: {}'
      .format(round(rmse_lstm, 3)))
```

```python
y_test_flat = y_test.flatten()
forecast_flat = forecast.flatten()
y_test_flat.shape, forecast_flat.shape
```

```python
plt.figure(figsize=(18,5))

plt.plot(y_test_flat[330:500], marker='.',label='Target')
plt.plot(forecast_flat[330:500], marker='.', label='Predictions')
plt.title('Elec consumption')
plt.xlabel('Time(h)')
plt.ylabel('Elec consump (khw)')
plt.legend()
plt.show()
```

```python
print( "The RMSE for Stacked LSTM model is: %f" %np.sqrt(mean_squared_error(y_test_flat,forecast_flat)))
print( "The accuracy of Stacked LSTM model is: %f" %r2_score(y_test_flat,forecast_flat))
```

```python
y_test_reverse = scaler.inverse_transform(y_test)
forecast_reverse = scaler.inverse_transform(forecast)

y_test_reverse.shape, forecast_reverse.shape
```

```python
y_test_reverse = y_test_reverse.flatten()
forecast_reverse = forecast_reverse.flatten()

y_test_reverse.shape, forecast_reverse.shape
```

```python
plt.figure(figsize=(18,5))

plt.plot(y_test_reverse[330:500], marker='.',label='Target')
plt.plot(forecast_reverse[330:500], marker='.', label='Predictions')
plt.title('Elec consumption')
plt.xlabel('Time(h)')
plt.ylabel('Elec consump (khw)')
plt.legend()
plt.show()
```
